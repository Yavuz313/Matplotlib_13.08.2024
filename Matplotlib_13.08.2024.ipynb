# Functional Method and Object Oriented Programing(OOP) Methods

### Functional Method

```python
plt.plot(age, salary_1)

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary By Age")
```

### Object Oriented Programing(OOP) Methods

```python
fig, ax = plt.subplots()

ax.plot(age, salary_1)
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary By Age")

plt.show()
```

### Functional Method(Coklu Grafik Cizme)

```python
plt.subplot(2, 1, 1) 
plt.plot(age, salary_1, "r")

plt.subplot(2, 1, 2)
plt.plot(age, salary_1, "b")

plt.show()
```

### OOP Method (Coklu Grafik Cizme)

```python
fig, ax = plt.subplots(2, 1,  figsize=(10,5), facecolor = "yellow") # 2 satir 1 sutun seklinde yazdirdik 

ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b")

plt.show()
```

```python
fig, ax = plt.subplots(1, 2) # 1 satir 2 sutun seklinde yazdirdik 

ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b")

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5), facecolor = "yellow") # 2 satir 1 sutun seklinde yazdirdik 
# Figure in icinde 2 tane grafik olusturduk 
ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b")

plt.show()
```

### OOP Method (Coklu Grafik Cizme) 2 

```python
fig, ax = plt.subplots(2,3,figsize=(10,5), facecolor = "yellow")

ax[0,0].plot(age, salary_1, "r")
ax[0,1].plot(age, salary_2, "b")
ax[0,2].plot(age, salary_1, "r")
ax[1,0].plot(age, salary_2, "b")
ax[1,1].plot(age, salary_1, "r")
ax[1,2].plot(age, salary_2, "b")

plt.tight_layout() # içiçe giren kısımları otomatik düzneler
plt.show()
```

```python
fig, ax = plt.subplots(2,3,figsize=(10,5), facecolor = "yellow")

ax[0][0].plot(age, salary_1, "r")
ax[0][1].plot(age, salary_1, "b")
ax[0][2].plot(age, salary_1, "r")
ax[1][0].plot(age, salary_1, "b")
ax[1][1].plot(age, salary_1, "r")
ax[1][2].plot(age, salary_1, "b")

plt.tight_layout() # içiçe giren kısımları otomatik düzneler
plt.show()
```

```python
x = np.arange(0, 30, 0.2)

fig,ax = plt.subplots(figsize=(12, 8))

ax.plot(x, x-1, color = "r", linewidth = 1, linestyle = "-")
ax.plot(x, x-2, color = "green", lw = 2, ls = "-.")
ax.plot(x, x-3, color = "cyan", lw = 3, ls = ":")
ax.plot(x, x-4, color = "m", lw = 4, ls = "--")

plt.show()
```

```python
x = np.arange(0, 10)

fig,ax = plt.subplots(2, 2, figsize=(12, 6))

ax[0, 0].plot(x, "r", marker = "+", markersize = "5")
ax[0, 1].plot(x-2, "r", marker = "o", markersize = "3")
ax[1, 0].plot(x-3, "blue", marker = "2", markersize = "4")
ax[1, 1].plot(x-4, "green", marker = "1", markersize = "5")

plt.show()
```

```python
x = np.arange(0, 10)
fig, ax = plt.subplots(2, 2, figsize=(12, 6))
ax[0, 0].plot(x, "r", marker = "+", markersize = "5")
ax[0, 0].set_title("Grafik - 1", fontsize = 18)
ax[0, 0].set_xlabel("X-Label", fontsize = 14)
ax[0, 0].set_ylabel("Y-Label", fontsize = 14)


ax[0, 1].plot(x-2, "y", marker = "o", markersize = "3")
ax[0, 1].set_title("Grafik - 2", fontsize = 18)
ax[0, 1].set_xlabel("X-Label", fontsize = 14)
ax[0, 1].set_ylabel("Y-Label", fontsize = 14)


ax[1, 0].plot(x-3, "blue", marker = "s", markersize = "5")
ax[1, 0].set_title("Grafik - 3", fontsize = 18)
ax[1, 0].set_xlabel("X-Label", fontsize = 14)
ax[1, 0].set_ylabel("Y-Label", fontsize = 14)


ax[1, 1].plot(x-4, "g", marker = "1", ms = "5")
ax[1, 1].set_title("Grafik - 4", fontsize = 18)
ax[1, 1].set_xlabel("X-Label", fontsize = 14)
ax[1, 1].set_ylabel("Y-Label", fontsize = 14)

plt.tight_layout() # içiçe giren kısımları otomatik düzneler
plt.show()
```
